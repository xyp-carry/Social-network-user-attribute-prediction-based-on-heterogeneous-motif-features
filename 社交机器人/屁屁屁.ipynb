```python
user_list = []
data = pd.read_csv('split.csv')
user_data = data[data.split != 'val']
for i in user_data['id']:
    user_list.append(i[1:])
```

```python
dict_user_tweet = {}
```

```python
def do_info(author_id, text_str, dict_info):
    dict_info[author_id]['num'] += 1
    if 'RT @' in text_str:
        dict_info[author_id]['RT'] += 1
    if r'https:' in text_str:
        dict_info[author_id]['URL'] += 1
    if len(re.findall('http', text_str)) > 1:
        dict_info[author_id]['UU'] += 1
    if '@' in text_str:
        dict_info[author_id]['@'] += 1
        if len(re.findall('@', text_str)) > 1:
            dict_info[author_id]['@@'] += 1


def do_hashtags(author_id, text_str, hashtags, dict_info):
    if len(hashtags) > 0:
        dict_info[author_id]['hashtag'] += 1
        a = 0
        for i in hashtags:
            a += len(re.findall(i, text_str))
        dict_info[author_id]['S_hashtag'] += a / len(hashtags)


def init_infodict(author, dict_info):
    dict_info[author] = {}
    dict_info[author]['num'] = 0
    dict_info[author]['RT'] = 0
    dict_info[author]['URL'] = 0
    dict_info[author]['UU'] = 0
    dict_info[author]['@'] = 0
    dict_info[author]['@@'] = 0
    dict_info[author]['hashtag'] = 0
    dict_info[author]['S_hashtag'] = 0
```

```python
stack = []
label = 1
b = -1  # 开头有一个'start_array' 所以初始化为-1
c = 0
v = 0
item_text = '0'
hashtags = []
with open('tweet_5.json', 'r', encoding='utf-8') as f:
    parser = ijson.parse(f)
    start = time.time()
    while 1:
        try:
            b += 1
            a = next(parser)
            # print(a)
            if time.time() - start > 1800:
                print(b)
                start = time.time()
        except StopIteration as e:
            print("完成")
            break
        except ijson.IncompleteJSONError as e:
            if b > 1000000:
                print(b)
                b = 0
            stack = []
            continue
        if a[1] == 'start_map' and a[0] == 'item':
            stack.append(1)
        if a[0] == 'item.author_id':
            # print(a)
            now_name = str(a[2])
            if str(now_name) in user_list:
                flag = 1
                if str(now_name) in dict_user_tweet.keys():
                    pass
                else:
                    init_infodict(now_name, dict_user_tweet)
            else:
                flag = 0
        if a[0] == 'item.entities.hashtags.item.text' and flag == 1:
                hashtags.append(a[2])
        if a[0] == 'item.text' and flag == 1:
            item_text = a[2]
        if item_text != '0':
            do_info(now_name, item_text, dict_user_tweet)
            if len(hashtags) > 0:
                do_hashtags(now_name, item_text,hashtags,dict_user_tweet)
        if a[1] == 'end_map' and len(stack) > 0 and a[0] == 'item':
            stack.pop()
            if len(stack) == 0 :
                item_text = '0'
                hashtags = []
                v += 1
                if v == 100000:
                    print('输出一条')
                    v = 0
```
